## 0. Introducción a Scikit-Learn

Existen varias librerias de Python que proveen implementaciones solidas de multiples algoritmos de aprendizaje de máquinas. Una de las más conocidas es [Scikit-Learn](http://scikit-learn.org). Scikit-Learn se caracteriza por ser una API limpia, uniforme y optimizada, y por tener una documentación en línea muy útil y completa. Una ventaja de esta uniformidad es que una vez que comprende el uso básico y la sintaxis de Scikit-Learn para un tipo de modelo, cambiar a un nuevo modelo o algoritmo es muy sencillo.

### Uso básico de la API

Por lo general, los pasos a seguir para utilizar los modelos implementados en Scikit-Learn son:

1. Seleccionar una clase de modelo importando la clase de estimador apropiada de Scikit-Learn.
2. Seleccionar los hiper parámetros del modelo al declarar la clase anterior con estos valores.
3. Organizar los datos en una matriz de variables y un vector objetivo.
4. Ajustar el modelo a los datos llamando al método `` .fit() `` de la instancia del modelo.
5. Predecir el modelo en nuevos datos.

## 1. Mínimos Cuadrados Ordinarios
### Regresión lineal univariada

Los modelos de aprendizaje supervisado buscan predecir o modelar una variable objetivo $y$ en función de ciertas características o atributos $x$ de cada observación

$$ y = f(x) $$

La función $f$ captura la manera en que las variables explicativas $x$ se combinan para obtener el resultado $y$. 

Los modelos de regresión lineal son un buen punto de partida para las tareas de regresión. Dichos modelos son populares porque pueden ajustarse muy rápidamente y son muy interpretables.

En el caso del modelo de Regresión Lineal, se supone que las variables se ponderan linealmente para construir $y$.

Un ajuste en línea recta es un modelo de la forma

$$
y = ax + b
$$
donde $a$ es la *pendiente*, y $b$ is el *intercepto*.

Considere los siguientes datos, que están dispersos alrededor de un modelo lineal con una pendiente de 2 y una intersección de -5:

```python
#!pip install spreg


# importar paquetes estandar
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import geopandas as gpd
import pandas as pd
import numpy as np
from shapely.geometry import Point
import shapely.ops as ops
import matplotlib.pyplot as plt
from sklearn import metrics
import libpysal as lps # Librería de python para análisis espacial
import spreg
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import RandomForestRegressor 
```

### Generar datos sintéticos

```python
# generamos puntos distribuidos de manera uniforme entre 0 y 10
# para tener replicabilidad
rng = np.random.RandomState(1)
x = 10 * rng.rand(50,1)
```

```python
rng.randn(50)
```

```python
# generamos los valores de y usando un modelo lineal con pendiente 2
# e intercepto -5 mas un ruido alatorio
y = 2 * x[:,0] - 5 + rng.randn(50)

# construit un grafico de puntos
plt.scatter(x, y);
```

```python
#### importar LinearRegression de Scikit-Learn
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression(fit_intercept = True)
```

```python
model.fit(x, y)
```

```python
print("Pendiente del modelo:    ", model.coef_[0])
print("Intercepto del modelo:", model.intercept_)
```

```python
model.predict(x)
```

```python
plt.scatter(x,y)
plt.plot(x, model.predict(x), color = 'red');
```

## Métricas de desempeño

- $MSE = \text{Mean squared error (error cuadrático medio)} = mean((Y-\hat Y)^2)$ 
<br>
<br>
- $MAE = \text{Mean absolute error (error absoluto medio)} = mean(\mid Y-\hat Y \mid)$ 
<br>
<br>
- $R^2 = \dfrac{\sigma^{2}_{XY}}{\sigma^{2}_{X}\sigma^{2}_{Y}}$ donde $\sigma^{2}_{XY}$ es la covarianza de $X,Y$, $\sigma^{2}_{X}$ la varianza de $X$ y $\sigma^{2}_{Y}$ la varianza de $Y$. 

```python
model.predict(np.array(0).reshape(-1,1))
```

```python
# realizar la predicción dentro de muestra
y_pred = model.predict(x)
```

```python
y_pred
```

#### La idea detrás de cualquier métrica o esquema de validación es comparar la predicción realizada por el modelo (y_pred, arriba) contra los datos reales conocidos de una muestra de datos (y).

La sintaxis de las métricas es, en general: importar la métrica deseada desde [sklearn.metrics](https://scikit-learn.org/stable/modules/classes.html#module-sklearn.metrics) y luego llamarla pasando los y_reales y los y_predichos.

```python
# error cuadrático medio en el entrenamiento
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y,y_pred)
print("El modelo de regresión lineal tiene un MSE = {0:.3f} dentro de muestra".format(mse))
```

```python
# error absoluto promedio en el entrenamiento
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y,y_pred)
print("El modelo de regresión lineal tiene un MAE = {0:.3f} dentro de muestra".format(mae))
```

```python
# R cuadrado en el entrenamiento
from sklearn.metrics import r2_score
r2 = r2_score(y,y_pred)
print("El modelo de regresión lineal tiene un R2 = {0:.3f} dentro de muestra".format(r2))
```

### Regresión Lineal Múltiple

El estimador ``LinearRegression`` también puede estimar modelos multivariables de la forma

$$
y = b_0 + b_1 x_1 + b_2 x_2 + \cdots + b_k x_k
$$

donde hay multiples valores para $x$.
Geométricamente, esto es ajustar un plano a puntos en tres dimenciones o un hiper plano para dimensiones mayores.

En términos analíticos, queremos que para unos mismos valores de $(b_0, b_1, b_2, ..., b_k)$, simultaneamente todos nuestros registros logren una buena predicción.

$$ y_1 = b_0 + b_1 x_{11} + b_2 x_{21} + \cdots + b_k x_{k1} $$
$$ y_2 = b_0 + b_1 x_{12} + b_2 x_{22} + \cdots + b_k x_{k2} $$
$$ \vdots $$
$$ y_n = b_0 + b_1 x_{1n} + b_2 x_{2n} + \cdots + b_k x_{kn} $$


Finalmente, la forma de estimar esto se conoce como Minimos Cuadrados Ordinarios y queremos que el error promedio que comete nuestro modelo sea el menor posible.

Filosóficamente, esto es lo que siempre se quiere en un modelo de aprendizaje de máquinas, y lo que varía es la forma en que se mide este error.

En este caso, queremos encontrar $\hat \beta$ que minimice el error $(Y-\hat Y)^2$ donde $\hat Y = X \hat \beta$. <br>

Esto se consigue con los coeficientes $\hat \beta = (X^TX)^{-1}X^TY$

```python
# creamos datos con 2 variables distribuidas uniformemente entre 0 y 10
# creamos la variable dependiente y como un modelo lineal de x
rng = np.random.RandomState(1234) # replicabilidad
x = 10 * rng.rand(100, 2)
y = 0.5 + np.dot(x, [1.5, -2.]) + rng.randn(100)
```

```python
modelo = LinearRegression(fit_intercept = True)
```

```python
modelo.fit(x, y)
```

```python
modelo.intercept_
```

```python
modelo.coef_
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import pandas as pd

fig = plt.figure(figsize=(20,20))
ax = fig.add_subplot(211,projection='3d')

ax.scatter(x.T[0],x.T[1], c='r', marker='o', s=5)

# generar una grilla de x,y
xx, yy = np.meshgrid(x.T[0],x.T[1])

exog = pd.core.frame.DataFrame({'X': xx.ravel(), 'Y': yy.ravel()})
out = modelo.predict(exog)
ax.plot_surface(xx, yy,
                out.reshape(xx.shape),
                rstride=1,
                cstride=1,
                color='None',
                alpha = 0.4)
plt.show()
```

## Zona comerciales

```python
comercios = pd.read_pickle('data/comercios.p', 'gzip')
comercios = comercios.reset_index(drop=True)
comercios = gpd.GeoDataFrame(comercios, geometry = comercios[['Longitude', 'Latitude']].apply(Point, axis=1), crs = {'init':'epsg:4326'})
```

## Definición Variable independiente

## Definición Variables dependientes

```python
modelo = {'modelo': modelo_ols,
         'datos': valor_ref}
```

```python
modelo_ols.predy
```

```python
metricas = ['RMSE: {:.3f}'.format(np.sqrt(metrics.mean_squared_error(valor_ref['V_REF'], modelo_ols.predy))),
            'MAE: {:.3f}'.format(metrics.mean_absolute_error(valor_ref['V_REF'], modelo_ols.predy))]

plt.figure(figsize = (12,8))
plt.scatter(valor_ref['V_REF'], modelo_ols.predy, s = 0.1, label = '\n'.join(metricas))
plt.plot([-0.2, 5e6], [-0.2, 5e6], color = 'black', linestyle = 'dotted')
plt.xlabel('Valor real')
plt.ylabel('Valor predicho')
plt.legend(loc = 'lower right')
plt.title('Valores reales y predichos metro cuadrado OLS \n Manzanas de entrenamiento en Bogotá', fontsize=16);
```

## Predictores laggeados (WX):

$$
\Large
\begin{equation} 
Valor_i = \alpha + \beta X_i + \delta \sum_j w_{ij} X'_i
\end{equation}
$$

```python
knnW = lps.weights.KNN.from_dataframe(valor_ref, k = 8)
knnW.transform = 'r'
```

```python
valor_ref['comercios_lag'] = lps.weights.lag_spatial(knnW, valor_ref['comercios'])
valor_ref['estrato_lag'] = lps.weights.lag_spatial(knnW, valor_ref['ESTRATO'])
```

```python
X = np.array(valor_ref[['comercios', 'comercios_sq', 'ESTRATO', 'comercios_lag', 'estrato_lag']])
```

```python
modelo_ols_lag = spreg.OLS(y, X, name_y = 'Valor m2', name_x = ['comercios', 'comercios_sq', 'ESTRATO', 'comercios_lag', 'estrato_lag'])
```

```python
print(modelo_ols_lag.summary)
```

```python
metricas = ['RMSE: {:.3f}'.format(np.sqrt(metrics.mean_squared_error(valor_ref['V_REF'], modelo_ols_lag.predy))),
            'MAE: {:.3f}'.format(metrics.mean_absolute_error(valor_ref['V_REF'], modelo_ols_lag.predy))]

plt.figure(figsize = (12,8))
plt.scatter(valor_ref['V_REF'], modelo_ols_lag.predy, s = 0.1, label = '\n'.join(metricas))
plt.plot([-0.2,5e6], [-0.2,5e6], color = 'black', linestyle = 'dotted')
plt.xlabel('Eventos reales')
plt.ylabel('Eventos predichos')
plt.legend(loc = 'lower right')
plt.title('Valores reales y predichos metro cuadrado OLS lag comercios \n Manzanas de entrenamiento en Bogotá', fontsize=16);
```

## Variable objetivo lag como predictor (WY)

$$
\Large
\begin{equation} 
Valor_i = \alpha + \beta X_i + \delta \sum_j w_{ij} X'_i + \lambda \sum_j w_{ij} Valor_i
\end{equation}
$$

```python
valor_ref['valor_lag'] = lps.weights.lag_spatial(knnW, valor_ref['V_REF'])
```

```python
X = np.array(valor_ref[['comercios', 'comercios_sq', 'ESTRATO', 'comercios_lag', 'estrato_lag', 'valor_lag']])
```

```python
modelo_s2sls_lag = spreg.OLS(y, X, name_y = 'Valor m2', name_x = ['comercios', 'comercios_sq', 'ESTRATO', 'comercios_lag', 'estrato_lag', 'valor_lag'])
```

```python
print(modelo_s2sls_lag.summary)
```

```python
metricas = ['RMSE: {:.3f}'.format(np.sqrt(metrics.mean_squared_error(valor_ref['V_REF'], modelo_s2sls_lag.predy))),
            'MAE: {:.3f}'.format(metrics.mean_absolute_error(valor_ref['V_REF'], modelo_s2sls_lag.predy))]

plt.figure(figsize = (12,8))
plt.scatter(valor_ref['V_REF'], modelo_s2sls_lag.predy, s = 0.1, label = '\n'.join(metricas))
plt.plot([-0.2,5e6], [-0.2,5e6], color = 'black', linestyle = 'dotted')
plt.xlabel('Eventos reales')
plt.ylabel('Eventos predichos')
plt.legend(loc = 'lower right')
plt.title('Valores reales y predichos metro cuadrado OLS lag comercios y valor \n Manzanas de entrenamiento en Bogotá', fontsize=16);
```
